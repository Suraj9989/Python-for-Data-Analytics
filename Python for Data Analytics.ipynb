```python
#Write a Program to build a calculator using function:
op = input("Enter operation (Addition, Subtraction, Multiply, Division, Modulus): ")

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

def Addition(a, b):
    print(a + b)

def Subtraction(a, b):
    print(a - b)

def Multiply(a, b):
    print(a * b)

def Division(a, b):
    print(a / b)

def Modulus(a, b):
    print(a % b)

if op == "Addition":
    Addition(x, y)

elif op == "Subtraction":
    Subtraction(x, y)

elif op == "Multiply":
    Multiply(x, y)

elif op == "Division":
    Division(x, y)

elif op == "Modulus":
    Modulus(x, y)

else:
    print("Invalid operation")
```

```python
def Addition(a, b):
    return a + b

def Subtraction(a, b):
    return a - b

def Multiply(a, b):
    return a * b

def Division(a, b):
    return a / b

def Modulus(a, b):
    return a % b

operations = {
    "Addition": Addition,
    "Subtraction": Subtraction,
    "Multiply": Multiply,
    "Division": Division,
    "Modulus": Modulus
}

opt = input("Enter operation: ")
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

if opt in operations:
    print(operations[opt](x, y))
else:
    print("Invalid operation")
```

```python
'''Write a program to find the grade earn by students
if marks is greater than equal to 90 gets 'A',
if marks is greater than equal to 75 gets 'B',
if marks is greater than equal to 50 gets 'C',
if marks is greater than equal to 33 gets 'D',
else 'Fail'
'''
marks = int(input("Enter Students marks"))
if (marks >= 90):
    print("Student gets 'A' grade")
else:
    if (marks >= 75):
        print("Student gets 'B' grade")
    else:
        if (marks >= 50):
            print("Student gets 'C' grade")
        else:
            if (marks >= 33):
                print("Student gets 'D' grade")
            else:
                print("Fail")
```
