```python
# Finding and Removing Outliers
df.boxplot('age')
```

```python
# boxplot is based on Inter Quantile Range
Q1 = df['age'].quantile(.25)
Q3 = df['age'].quantile(.75)

IQR = Q3 - Q1

LL = Q1-1.5*IQR
UL = Q3+1.5*IQR


median = df['age'].median()

# np.where(condition, value_if_true, value_if_false)

df['age'] = np.where( df['age'] > UL , median, df['age'])
df['age'] = np.where( df['age'] < LL , median, df['age'])
```

```python
# 1. Create age groups like 18-25, 26-35, 36-45, 46-55, 56-65, 65+. 
# For each age group, calculate the percentage of people earning >50K and visualize it. 
# Hint: pd.cut(), groupby(), apply() / mean(), plot(kind='bar')
# Create age groups
df['Age_Group'] = pd.cut(
    df['age'],
    bins=[18, 25, 35, 45, 55, 65, float('inf')],
    labels=['18-25', '26-35', '36-45', '46-55', '56-65', '65+'],
    include_lowest=True
)

# Calculate percentage earning >50K in each age group
income_pct = (
    df.groupby('Age_Group')['income']
      .apply(lambda x: (x == '>50K').mean() * 100)
)

# Display percentages
print(income_pct)

# Visualize
income_pct.plot(
    kind='bar',
    title='Percentage of People Earning >50K by Age Group',
    ylabel='Percentage (%)',
    xlabel='Age Group'
)
plt.show()
```

```python
# 2. Find the top 10 occupations where the percentage of people earning >50K is highest. Ignore occupations with fewer than 100 records. 
# Hint: groupby(), agg(), filtering, sort_values(), head(), plot(kind='barh')

# df.groupby('column').agg(
#     new_column1=('existing_column', 'function'),
#     new_column2=('existing_column', another_function)
# )

result = (
    df.groupby('occupation')
      .agg(
          total_records=('income', 'count'),
          income_pct=('income', lambda x: (x == '>50K').mean() * 100)
      )
)
print(result)

# Keep only occupations with at least 100 records
result = result[result['total_records'] >= 100]
print(result)
print()

# Top 10 occupations
top10 = result.sort_values('income_pct', ascending=False).head(10)
print(top10)
# Visualize
top10.sort_values('income_pct').plot(
    kind='barh',
    y='income_pct',
    figsize=(8, 5),
    title='Top 10 Occupations by % of People Earning >50K'
)
plt.show()
```

```python
# 3. Compare the average capital-gain across different education levels and visualize the top 10 education categories. 
# Hint: groupby(), mean(), sort_values(), head(), plot(kind='bar')

result = df.groupby('education')['capital-gain'].mean()
print(result)
result = result.sort_values(ascending=False)
result.head(10).plot(
    kind='bar',
    title=("Avarage capital gain by Education"),
    figsize=(8, 5),
)
plt.show()
```

```python
# 4. Create separate histograms for hours-per-week for people earning <=50K and >50K. 
# Hint: filtering, boolean indexing, plot(kind='hist'), alpha, legend

plt.figure(figsize=(8,5))

df[df['income'] == '<=50K']['hour-per-week'].plot(
    kind='hist',
    bins=20,
    alpha=0.5,
    label='<=50K'
)

df[df['income'] == '>50K']['hour-per-week'].plot(
    kind='hist',
    bins=20,
    alpha=0.5,
    label='>50K'
)

plt.legend()
plt.xlabel('Hours per Week')
plt.show()
```

```python
# 5. Find the top 10 native countries with the highest percentage of people earning >50K. 
# Consider only countries with at least 50 records. 
# Hint: groupby(), agg(), filtering, sort_values(), head(), plot(kind='bar')

result1 = (
    df.groupby('native-country')
    .agg(
        total_records=('income', 'count'),
        income_pct1=('income', lambda x: (x == '>50K').mean() * 100)
    )
)
result1 = result1[result1['total_records'] >= 50]
top10 = result1.sort_values('income_pct1', ascending=False).head(10)
top10['income_pct1'].plot(
    kind = 'bar'
)
plt.show()
```

```python
# 6. Find the average hours-per-week for each marital-status category and visualize it from highest to lowest. 
# Hint: groupby(), mean(), sort_values(), plot(kind='bar'))

a = df.groupby('marital-status')['hour-per-week'].mean()
b = a.sort_values(ascending=False)

b.plot(
    kind= 'bar'
)
plt.show()
```

```python
# 7. For each relationship category, show the count of people earning <=50K and >50K using a stacked bar chart. 
# Hint: pd.crosstab(), plot(kind='bar', stacked=True)

ct = pd.crosstab(df['relationship'], df['income'])

print(ct)

ct.plot(
    kind='bar',
    stacked=True,
    figsize=(8, 5),
    title='Income Distribution by Relationship Status'
)

plt.xlabel('Relationship')
plt.ylabel('Count')
plt.show()
```

```python
# 8. For each race category, calculate the percentage of people earning >50K and visualize it. 
# Hint: groupby(), apply() / mean() lambda condition, plot(kind='bar')

a = ( df.groupby('race')['income']
     .apply(lambda x: (x == '>50K').mean() * 100)
    )

a.plot(
    kind='bar'
)
plt.show()
```
